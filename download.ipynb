```python
import numpy as np 
import pandas as pd 
import sys, requests, shutil, os
from urllib import request, error

def download_image(url):
    response = requests.get(url, stream=True)
    with open('./train_images/image.jpg', 'wb') as out:
        shutil.copyfileobj(response.raw, out)
    del response

data = pd.read_csv('train_sample.csv')
```

```python
urls = data['url'].dropna()
data.head(10)
```

```python
for url in urls:
    if url !='None':
        img_id = data.loc[data['url'] == url, 'id'].iloc[0]
    
        if os.path.exists('./train_images/'+str(img_id)+'.jpg'):
            continue
        else:
            try:
                download_image(url)
                os.rename('./train_images/image.jpg', './train_images/' + str(img_id) + '.jpg')
            except:
                continue
```
